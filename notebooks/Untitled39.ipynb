## Model Training

```python
df = pd.read_csv("/content/Algerian_forest_fires_cleaned_dataset.csv")
df.head()
```

```python
df.columns
```

```python
## Encoding of the categories in the classes
df = df.drop(['day', 'month', 'year'], axis = 1)
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 0, 1)
```

```python
df['Classes'].value_counts()
```

```python
df.head()
```

```python
## Dependent and Independent features

X = df.drop('FWI', axis=1)
y=df['FWI']
```

```python
X.head()
```

```python
## Train-test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
## Feature selection based on correlation

X_train.corr()
```

```python
## Check for multicollinearity --> Pearson correlation

## The purpose behind finding correlation is that we are trying to find the
## features which are highly correlated so that we can easily remove those
## features because they those features are same so we dont have to take two separate features.
## we can drop one of the feature.


## In multicollinearity --> the independent features are highly correlated

plt.figure(figsize=(12, 10))

corr = X_train.corr()
sns.heatmap(corr, annot=True)
```

```python
def correlation(dataset, threshold):
  col_corr = set()
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i, j]) > threshold:
        colname = corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```

```python
## Threshold domain expertise

corr_features = correlation(X_train, 0.85)
```

```python
## Drop features when correlation is more than 0.85

X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
X_train.shape, X_test.shape
```

## Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

linear = LinearRegression()
linear.fit(X_train_scaled, y_train)
y_pred = linear.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean Absolute Error: ", mae)
print("R2 Score: ", score)
plt.scatter(y_test, y_pred)
```

### Explanation of `alpha_` in ElasticNetCV

In regularized linear models like Lasso, Ridge, and ElasticNet, a parameter called `alpha` (or lambda) controls the strength of the regularization. Regularization is a technique used to prevent overfitting by adding a penalty to the loss function. This penalty discourages the model from assigning excessive weights to features.

For `ElasticNet` specifically, there are two main hyperparameters: `alpha` and `l1_ratio`.

*   **`alpha` (regularization strength)**: This parameter controls the overall strength of the penalty. A higher `alpha` value means stronger regularization, leading to simpler models and potentially preventing overfitting. A lower `alpha` means less regularization, allowing the model to fit the training data more closely.
*   **`l1_ratio` (mixing parameter)**: This parameter controls the mix between L1 (Lasso) and L2 (Ridge) regularization. An `l1_ratio` of 1 corresponds to Lasso regression, while an `l1_ratio` of 0 corresponds to Ridge regression. Values between 0 and 1 represent a combination of both.

#### How `ElasticNetCV` works:

`ElasticNetCV` is a version of ElasticNet that uses cross-validation to automatically find the best `alpha` (and `l1_ratio` if not explicitly set) for your model. Here's how it generally works:

1.  **Multiple `alpha` values**: `ElasticNetCV` tries a range of `alpha` values (and `l1_ratio` values, if `l1_ratio` is given as a list of possibilities) during the training process.
2.  **Cross-validation**: For each combination of `alpha` (and `l1_ratio`), it performs cross-validation on the training data. This involves splitting the training data into several folds, training the model on a subset of the folds, and evaluating it on the remaining fold. This process is repeated for all folds.
3.  **Optimal `alpha_`**: After evaluating the model's performance (e.g., using mean squared error as the scoring metric) for all `alpha` (and `l1_ratio`) values across all cross-validation folds, `ElasticNetCV` selects the `alpha` that yields the best average performance. This chosen optimal `alpha` is then stored in the `alpha_` attribute.

So, `enetcv.alpha_` represents the regularization strength (`alpha`) that the `ElasticNetCV` model determined to be the most effective for your dataset through its internal cross-validation process, given the default `l1_ratio` values it explores. It's the `alpha` value that should theoretically lead to the best generalization performance.
